```python
import getpass
import os

if not os.environ.get("GOOGLE_API_KEY"):
  os.environ["GOOGLE_API_KEY"] = "YOUR_GOOGLE_API_KEY"

from langchain.chat_models import init_chat_model

model = init_chat_model("gemini-2.0-flash", model_provider="google_genai")
```

```python
message = model.invoke("Who is the prime minister of India?")
print(message)

message = model.invoke("What is LangChain?")
print(message)
```

```python
from langchain_core.messages import HumanMessage, SystemMessage
message = [
    SystemMessage("Translate the following from English into Hindi"),
    HumanMessage("Hey! How are you?"),
]

model.invoke(message)
```

```python
model.invoke("hello")
model.invoke([{"role": "user", "content":"Hello"}])
model.invoke([HumanMessage("Hello")])
```

Prompt Template

```python
from langchain_core.prompts import ChatPromptTemplate
system_template = "Translate the following from english to {language}"
prompt_template = ChatPromptTemplate.from_messages(
    [("system", system_template),("user","{text}")]
)
```

```python
prompt = prompt_template.invoke({"language": "Japanese", "text": "How are you?"})
print(prompt)
```

```python
prompt.to_messages()
```

```python
response = model.invoke(prompt)
print(response.content)
```
