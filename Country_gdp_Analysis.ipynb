# what is univariate?  
  #### Plot the graph using 1 variable is called univariate analysis.

```python
vis2 = sns.displot(df["InternetUsers"])
```

```python
vis3 = sns.distplot(df["InternetUsers"], bins = 15) # bins used to create number of graph box.
```

```python
plt.rcParams['figure.figsize'] = 8,7
```

```python
vis4 = sns.boxplot(data = df, x = "IncomeGroup", y = "BirthRate")
```

# what is bivariate?
#### Plot the graph using 2 variable is called bivariate analysis.

```python
vis5 = sns.lmplot(data = df, x = "InternetUsers", y = "BirthRate")
```

```python
vis5 = sns.lmplot(data = df, x = "InternetUsers", y = "BirthRate", fit_reg = False) #  fit_reg can add or remove the straight line present in the graph. by using True or False.
```

```python
vis5 = sns.lmplot(data = df, x = "InternetUsers", y = "BirthRate", fit_reg = False, hue = 'IncomeGroup')
```

```python
vis5 = sns.lmplot(data = df, x = "InternetUsers", y = "BirthRate", fit_reg = True, hue = 'IncomeGroup')
```
