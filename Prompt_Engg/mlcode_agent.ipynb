```python
os.environ["GOOGLE_API_KEY"] = " "
genai.configure(api_key=os.getenv("GOOGLE_API_KEY"))
```

```python
model = LiteLLMModel(
    model_id="gemini/gemini-2.0-flash", 
    api_key=os.getenv("GOOGLE_API_KEY")
)
```

```python
ml_code_agent = CodeAgent(
    tools = [DuckDuckGoSearchTool()],
    additional_authorized_imports=['pandas', 'numpy', 'sklearn', 'json'],
    model=model
)
```

```python
def generate_ml_code(prompt):
    full_prompt = """ 
    create a step-by-step plan to write python code for svm model usig sklearn to classify the iris dataset"""
    try:
        response = ml_code_agent.run(full_prompt)
        return response
    except Exception as e:
        return f"Error: {str(e)}"
```

```python
ifacce = gr.Interface(
    fn = generate_ml_code, 
    inputs="text",
    outputs= "text",
    title= "ML code Generator(Gemini 2.0 flash)",
    description="Enter a ML Code "
)
```

```python
ifacce.launch()
```
